```python
import pandas as pd, numpy as np, sklearn,matplotlib.pyplot as plt, seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("../data/CleanData/fullData.csv")
data = data[["YEAR FORECAST MADE",'YEAR BEING FORECAST',"FORECASTER_CONF", 'INDICATOR', 'actual', 'pred_average', 'pred_var', 'banana', 'beef', 'bread', 'chicken','eggs', 'electricity', 'flour', 'iceCream','unleadedGasoline']].dropna()
train = data[data["YEAR BEING FORECAST"]<2017]
val = data[(data["YEAR BEING FORECAST"]==2017)|(data["YEAR BEING FORECAST"]==2018)]
test = data[data["YEAR BEING FORECAST"]==2019]
```

```python
train
```

```python
def MeasureSplit(df,measure,year):
    df = df[df["YEAR BEING FORECAST"]<=year]
    return df[df["INDICATOR"]==measure].drop(columns = ["actual","INDICATOR","YEAR BEING FORECAST","YEAR FORECAST MADE"]),df[df["INDICATOR"]==measure]["actual"]
```

```python
model = RandomForestRegressor()
```

```python
X_train,y_train = MeasureSplit(train,"RealGDP")
model.fit(X_train,y_train)
```

```python
X_val,y_val = MeasureSplit(val,"RealGDP")
X_test,y_test = MeasureSplit(test,"RealGDP")
```

```python
rmse_val = np.sqrt(np.mean(y_val-model.predict(X_val))**2)
```

```python
rmse_val
```

```python
r2_val = r2_score(y_val,model.predict(X_val))
r2_val
```

```python
rmse_spf = np.sqrt(np.mean(y_val-X_val["pred_average"].values)**2)
rmse_spf
```

```python
rmse_test = np.sqrt(np.mean(y_test-model.predict(X_test))**2)
rmse_test
```

```python
rmse_spf_test = np.sqrt(np.mean(y_test-X_test["pred_average"].values)**2)
rmse_spf_test
```

```python
model_scores=[]
model_predictions = []
for year in np.arange(1992,2019):
    X_train,y_train = MeasureSplit(data,"RealGDP",year)
    model = RandomForestRegressor()
    model.fit(X_train,y_train)
    X_test = data[(data["INDICATOR"]=="RealGDP")&(data["YEAR BEING FORECAST"]==year+1)].drop(columns = ["actual","INDICATOR","YEAR BEING FORECAST","YEAR FORECAST MADE"])
    y_test = data[(data["INDICATOR"]=="RealGDP")&(data["YEAR BEING FORECAST"]==year+1)]["actual"]
    predictions = model.predict(X_test)
    model_predictions.append(predictions)
    model_scores.append(np.sqrt(np.mean(y_test-predictions)**2))
```

```python
np.mean(model_scores)
```

```python
spf_scores=[]
for year in np.arange(1992,2019):
    X = data[(data["INDICATOR"]=="RealGDP")&(data["YEAR BEING FORECAST"]==year+1)].drop(columns = ["actual","INDICATOR","YEAR BEING FORECAST","YEAR FORECAST MADE"])
    y = data[(data["INDICATOR"]=="RealGDP")&(data["YEAR BEING FORECAST"]==year+1)]["actual"]
    spf_scores.append(np.sqrt(np.mean(y-X["pred_average"])**2))
```
