```python
import pandas as pd
import numpy as np

data = pd.read_csv("../data/CleanData/fullData.csv")#,index_col=0) 
```

```python
datagrouped = data.groupby(["FORECASTER ID", "INDICATOR"]).agg({"ACTUAL_CONF": np.mean})
```

## First Version - Gets average confidence for forecaster by Indicator

```python
datagrouped = datagrouped.reset_index()
datagrouped
```

## Second Version - Average Confidence Overall

```python
datagrouped2 = data.groupby(["FORECASTER ID"]).agg({"ACTUAL_CONF": np.mean})
```

```python
datagrouped2 = datagrouped2.reset_index()
datagrouped2
```

## Testing Merge

```python
relevant = data[data["YEAR FORECAST MADE"]>=1992][["YEAR FORECAST MADE",'YEAR BEING FORECAST', 'FORECASTER ID', 'INDICATOR', 'actual', 'MAX', 'pred_average', 'pred_var', 'banana', 'beef', 'bread', 'chicken',
        'eggs', 'electricity', 'flour', 'iceCream',
       'unleadedGasoline']]
```

```python
predict_set = relevant[relevant["YEAR BEING FORECAST"]>2021]
```

```python
predict_set.head()
```

```python
# First Way Merge
predicted_conf = pd.merge(predict_set, datagrouped, how="left", left_on=["FORECASTER ID", "INDICATOR"], right_on=["FORECASTER ID", "INDICATOR"] )
```

```python
predicted_conf.head()
```

```python
# Second Way Merge
predicted_conf2 = pd.merge(predict_set, datagrouped2, how="left", left_on=["FORECASTER ID"], right_on=["FORECASTER ID"] )
```
